```python
# *** Parte 3 - Processamento de Texto ***

# Instalação e downloads necessários (apenas no Colab)
!pip install -q nltk spacy
!python -m spacy download pt_core_news_sm

import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('rslp')

import re
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
import string
from collections import Counter
import spacy

# Texto base
texto = """
A alma é, pois, imortal; renasceu repetidas vezes na existência e contemplou todas as coisas existentes
e por isso não há nada que ela não conheça! Não é de espantar que ela seja capaz de evocar à memória a
lembrança de objetos que viu anteriormente, e que se relacionam tanto com a virtude como com as outras
coisas existentes. Toda a natureza, com efeito, é uma só, é um todo orgânico, e o espírito já viu todas
as coisas; logo, nada impede que ao nos lembrarmos de uma coisa – o que nós, homens, chamamos de “saber” –
todas as outras coisas acorram imediata e maquinalmente à nossa consciência.
"""
```

```python
# Questão 8 (Bag-of-Words (BoW)):
# Usa regex para tokenização simples e evita dependências problemáticas do NLTK

tokens = re.findall(r'\b\w+\b', texto.lower())
tokens_filtrados = [t for t in tokens if t.isalpha()]
bow = Counter(tokens_filtrados)

df_bow = pd.DataFrame(bow.items(), columns=['Palavra', 'Frequência']) \
    .sort_values(by='Frequência', ascending=False) \
    .reset_index(drop=True)

# Exibir todas as linhas
pd.set_option('display.max_rows', None)

print("Bag-of-Words (vetorização):")
print(df_bow)
print()
```

```python
# Questão 9 (Bag-of-n-Grams):

def gerar_ngrams(lista_tokens, n):
    """Gera lista de n-grams a partir da lista de tokens"""
    return list(zip(*[lista_tokens[i:] for i in range(n)]))

unigrams = gerar_ngrams(tokens_filtrados, 1)
bigrams = gerar_ngrams(tokens_filtrados, 2)
trigrams = gerar_ngrams(tokens_filtrados, 3)

print("Número de Unigrams:", len(unigrams))
print("Número de Bigrams:", len(bigrams))
print("Número de Trigrams:", len(trigrams))
print()
```

```python
# Questão 10 (Remoção de Stopwords):

stopwords_pt = set(stopwords.words('portuguese'))
tokens_sem_stopwords = [t for t in tokens_filtrados if t not in stopwords_pt]

print("Texto sem stopwords:")
print(" ".join(tokens_sem_stopwords))
print()
```

```python
# Questão 11 (Stemming com RSLPStemmer):

stemmer = RSLPStemmer()
tokens_stem = [stemmer.stem(t) for t in tokens_sem_stopwords]

print("Texto após stemming:")
print(" ".join(tokens_stem))
print()
```

```python
# Questão 12 (Lemmatization com spaCy):
# NLTK não possui lemmatizador em português nativo, por isso spaCy é usado aqui

nlp = spacy.load("pt_core_news_sm")
texto_filtrado = " ".join(tokens_sem_stopwords)
doc = nlp(texto_filtrado)

tokens_lematizados = [token.lemma_ for token in doc]

print("Texto após lematização (spaCy):")
print(" ".join(tokens_lematizados))
```
