```python
# *** Parte 1 - Normalização de Features ***
# Questão 1 (StandardScaler - sklearn):

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Carrega o dataset Iris
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Exibe os dados originais
print("Dados originais:")
display(df_iris.describe())

# Aplica a normalização com StandardScaler
scaler = StandardScaler()
dados_normalizados = scaler.fit_transform(df_iris)

# Cria DataFrame com os dados escalonados
df_normalizado = pd.DataFrame(dados_normalizados, columns=df_iris.columns)

# Exibe os dados transformados
print("\nDados normalizados com StandardScaler:")
display(df_normalizado.describe())

# Calcula e exibe a média e o desvio padrão
print("\nMédia das features (esperado ≈ 0):")
print(df_normalizado.mean())

print("\nDesvio padrão das features (esperado ≈ 1):")
print(df_normalizado.std())
```

```python
# Questão 2 (StandardScaler manual):

import numpy as np

# Função que aplica o escalonamento padrão manual
def escalonamento_padrao_manual(dados):
    medias = np.mean(dados, axis=0)
    desvios = np.std(dados, axis=0)
    return (dados - medias) / desvios

# Aplica a função ao dataset original
dados_np = df_iris.values
dados_escalonados_manualmente = escalonamento_padrao_manual(dados_np)

# Cria DataFrame com os dados manualmente escalonados
df_manual = pd.DataFrame(dados_escalonados_manualmente, columns=df_iris.columns)

# Exibe os dados normalizados manualmente
print("Dados normalizados manualmente:")
display(df_manual.describe())

# Compara os resultados com o StandardScaler
comparacao = np.allclose(df_normalizado, df_manual, atol=1e-8)
print("\nOs resultados manuais são equivalentes ao StandardScaler?", comparacao)
```

```python
# Questão 3 (Normalização L2 - Norma Euclidiana)

from sklearn.preprocessing import Normalizer

# Aplica normalização L2 (linha com norma Euclidiana igual a 1)
normalizador_l2 = Normalizer(norm='l2')
dados_l2 = normalizador_l2.fit_transform(df_iris)

# Cria DataFrame com os dados normalizados
df_l2 = pd.DataFrame(dados_l2, columns=df_iris.columns)

print("Dados normalizados com norma L2:")
display(df_l2.head())

# Verifica a norma L2 das primeiras linhas
normas_l2 = np.linalg.norm(dados_l2, ord=2, axis=1)
print("\nNorma L2 das primeiras 5 amostras:")
print(normas_l2[:5])
```

```python
# Questão 4 (Normalização L1 - Norma Manhattan)

# Aplica normalização L1 (linha com soma absoluta igual a 1)
normalizador_l1 = Normalizer(norm='l1')
dados_l1 = normalizador_l1.fit_transform(df_iris)

# Cria DataFrame com os dados normalizados
df_l1 = pd.DataFrame(dados_l1, columns=df_iris.columns)

print("Dados normalizados com norma L1:")
display(df_l1.head())

# Verifica a soma L1 das primeiras linhas
somas_l1 = dados_l1.sum(axis=1)
print("\nSoma L1 das primeiras 5 amostras (esperado = 1):")
print(somas_l1[:5])
```

```python
# *** Parte 2 - Seleção de Features ***
# Questão 5 (Estratégia de filtragem com Breast Cancer Dataset):

from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np

# Carrega o dataset
dados = load_breast_cancer()
X = pd.DataFrame(dados.data, columns=dados.feature_names)
y = dados.target

# Calcula a correlação de Pearson entre cada feature e a variável alvo
correlacoes = X.corrwith(pd.Series(y))

# Ordena as correlações em ordem decrescente de relevância
correlacoes_ordenadas = correlacoes.abs().sort_values(ascending=False)

# Seleciona as 10 features mais correlacionadas
top_10_features = correlacoes_ordenadas.head(10).index.tolist()
X_filtrado = X[top_10_features]

print("Top 10 features selecionadas por filtragem (correlação com o target):")
for i, feature in enumerate(top_10_features, start=1):
    print(f"{i}. {feature}")
```

```python
# Questão 6 (Método Wrapper (RFE)):

from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Cria modelo base
modelo = LogisticRegression(max_iter=5000)

# Aplica RFE (Recursive Feature Elimination)
rfe = RFE(estimator=modelo, n_features_to_select=10)
rfe.fit(X, y)

# Coleta as features selecionadas
features_wrapper = X.columns[rfe.support_].tolist()

print("Top 10 features selecionadas pelo método Wrapper (RFE):")
for i, feature in enumerate(features_wrapper, start=1):
    print(f"{i}. {feature}")
```

```python
# Questão 7 (Método Embedding):

from sklearn.linear_model import LassoCV

modelo_lasso = LassoCV()
modelo_lasso.fit(X, y)

# Coleta features com coeficiente diferente de zero
features_embedding = X.columns[modelo_lasso.coef_ != 0].tolist()

print("Features selecionadas pelo método Embedding (Lasso):")
for i, feature in enumerate(features_embedding, start=1):
    print(f"{i}. {feature}")
```
